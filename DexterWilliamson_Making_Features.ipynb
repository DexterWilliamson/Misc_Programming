```python
# Wrote a function that can do what I have written above to any 
# string that is passsed to it.

def int_rate_to_float(cell_contents):

  return float(cell_contents.strip().strip('%'))
```

```python
# Creating some examples that represent the cases that I want to clean up
examples = ['owner', 'Supervisor', ' Project Manager', np.NaN]
```

```python
# Wrote a function to clean up these use cases and increase uniformity.

def clean_title(job_title):

  if isinstance(job_title, str):
    return job_title.title().strip()
  else:
    return "Unknown"

for title in examples:
  print(clean_title(title))  
```

```python
# list comprehensions combine function calls and for loops over lists
# into one succinct and fairly readable single line of code.
[clean_title(example) for example in examples]
```

```python
# A function that works as expected. applied it to a column.
# This time it will overwrite the original
df['emp_title'] = df['emp_title'].apply(clean_title)
```

```python
#Create a function so i have a reusable piece of code that also allows me to use .apply()
#This stips away month and any blank spaces while checking if term is a string.
#It then returns an int 
def term_to_int(term):

  if isinstance(term, str):

    return int(term.strip('months').strip())

  else:

    return 0

df['term'] = df['term'].apply(term_to_int)
```

```python
#Checking if it worked
df['term'].head(20)
```

```python
#Function that checks whether loan is either Current/Fully Paid or not.
#Returns 1 if so or 0 if not.
def loan_status_bool(loan):

  if loan == 'Current' or loan == 'Fully Paid':

    return 1

  else:

    return 0

df['loan_status_is_great'] = df['loan_status'].apply(loan_status_bool)
```
